```python
df = water

# 'u'와 'v' 열의 분포 시각화
plt.figure(figsize=(12, 6))  # 그래프의 크기 설정

# 'u' 열의 히스토그램
plt.subplot(1, 2, 1)
sns.histplot(df['u'], bins=30, kde=True)
plt.title('Real water u')
plt.xlabel('u')

# 'v' 열의 히스토그램
plt.subplot(1, 2, 2)
sns.histplot(df['v'], bins=30, kde=True)
plt.title('Real water v')
plt.xlabel('v')

plt.tight_layout()
plt.savefig("image/real_water.png", transparent=True)
plt.show()
```

```python
df = wind

# 'u'와 'v' 열의 분포 시각화
plt.figure(figsize=(12, 6))  # 그래프의 크기 설정

# 'u' 열의 히스토그램
plt.subplot(1, 2, 1)
sns.histplot(df['u'], bins=30, kde=True)
plt.title('Real wind u')
plt.xlabel('u')

# 'v' 열의 히스토그램
plt.subplot(1, 2, 2)
sns.histplot(df['v'], bins=30, kde=True)
plt.title('Real wind v')
plt.xlabel('v')

plt.tight_layout()
plt.savefig("image/real_wind.png", transparent=True)

plt.show()
```

```python
df1 = water
df2 = wind

# Visualize the distribution of 'u' and 'v' columns with boxplots
plt.figure(figsize=(12, 12))  # Set the size of the graph

# Boxplot of column 'u'
plt.subplot(2, 2, 1)
sns.boxplot(data=df1, y='u')
plt.title('Real water u')
plt.ylabel('real water u')

# Boxplot of column 'v'
plt.subplot(2, 2, 2)
sns.boxplot(data=df1, y='v')
plt.title('Real water v')
plt.ylabel('water v')

# Boxplot of column 'u'
plt.subplot(2, 2, 3)
sns.boxplot(data=df2, y='u')
plt.title('Real wind u')
plt.ylabel('wind u')

# Boxplot of column 'v'
plt.subplot(2, 2, 4)
sns.boxplot(data=df2, y='v')
plt.title('Real wind v')
plt.ylabel('wind v')

plt.tight_layout()
plt.show()
```
