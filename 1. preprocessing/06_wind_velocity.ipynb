```python
import pandas as pd
import netCDF4 as nc
```

```python
# NetCDF 파일 열기
file_path = "data/wind_velocity.nc"
dataset = nc.Dataset(file_path, "r")
```

```python
# 파일 내의 변수 이름 확인
variables = dataset.variables.keys()
print(variables)
```

```python
t = [i for i in range(int(0 * 100), int((729.75 + 0.25) * 100), int(0.25 * 100)) for _ in range(8648)]
t = [x / 100 for x in t]

lati = []
for i in range(len(lat)):
    cnt = [lat.iloc[i][0]] * 92
    lati += cnt
lati *= 2920

long = []
for i in range(len(lon)):
    long.append(lon.iloc[i][0])
long *= 94
long *= 2920

u_lst = []
for i in range(2920):
    x = pd.DataFrame(dataset["u10"][i][:])
    new_x = x.values.flatten().tolist()
    u_lst.append(new_x)

v_lst = []
for i in range(2920):
    x = pd.DataFrame(dataset["v10"][i][:])
    new_x = x.values.flatten().tolist()
    v_lst.append(new_x)

def flatten(lst):
    flattened_lst = []
    for sublist in lst:
        for item in sublist:
            flattened_lst.append(item)
    return flattened_lst

u_lst = flatten(u_lst)
v_lst = flatten(v_lst)
```

```python
print(len(t))
print(len(lati))
print(len(long))
print(len(u_lst))
print(len(v_lst))
```

```python
data = {
    "time" : t,
    "longitude" : long,
    "latitude" : lati,
    "u" : u_lst,
    "v" : v_lst
}
columns = ["time","longitude","latitude","u","v"]
total = pd.DataFrame(data, columns = columns)
total
```
